```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
# Scales the features of the 'data_training' dataset to the range [0, 1] and stores the scaled data as an array in 'data_training_array'.
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
data_training_array.shape
```

```python
x_train = []
y_train = []

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100: i])
    y_train.append(data_training_array[i, 0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# ML model
```

```python
!pip install keras
```

```python
!pip install tensorflow
```

```python
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()
model.add(LSTM(units = 50, activation = 'relu', return_sequences = True,
               input_shape = (x_train.shape[1], 1)))
model.add(Dropout(0.2))

model = Sequential()
model.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
model.add(Dropout(0.3))

model = Sequential()
model.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
model.add(Dropout(0.4))

model = Sequential()
model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))
```

```python
# Initialize a Sequential model
model = Sequential()

# Add a LSTM layer with 50 units, ReLU activation, and returning sequences; specify input shape
model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_train.shape[1], 1)))

# Add a Dropout layer to prevent overfitting; 20% of neurons will be ignored during training
model.add(Dropout(0.2))

# Add another LSTM layer with 60 units, ReLU activation, and returning sequences
model.add(LSTM(units=60, activation='relu', return_sequences=True))

# Add a Dropout layer to prevent overfitting; 30% of neurons will be ignored during training
model.add(Dropout(0.3))

# Add another LSTM layer with 80 units, ReLU activation, and returning sequences
model.add(LSTM(units=80, activation='relu', return_sequences=True))

# Add a Dropout layer to prevent overfitting; 40% of neurons will be ignored during training
model.add(Dropout(0.4))

# Add a final LSTM layer with 120 units and ReLU activation
model.add(LSTM(units=120, activation='relu'))

# Add a final Dropout layer to prevent overfitting; 50% of neurons will be ignored during training
model.add(Dropout(0.5))

model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=50)
```

```python
model.save('my_model.keras')
```
